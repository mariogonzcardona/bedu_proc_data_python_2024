```python
# Obtenemos un panorama general de los datos.
def info_general_df(df):
  print('Información general del DataFrame')
  print(f'Cantidad de filas y columnas: {df.shape}')
  print("-"*50)
  print(f'Cantidad de datos nulos por columna: \n{df.isnull().sum()}')
  print("-"*50)
  print(f'Cantidad de datos únicos por columna: \n{df.nunique()}')
  print("-"*50)
  print(f'Tipos de datos por columna: \n{df.dtypes}')

info_general_df(df)
```

```python
# Generamos una copia del dataframe original.
df_current = df.copy()
df_current.head()
```

```python
# Convertimos los las columnas en sus tipos de datos correspondientes.
df_current['DOJ'] = pd.to_datetime(df_current['DOJ'], errors='coerce')
df_current['CURRENT DATE'] = pd.to_datetime(df_current['CURRENT DATE'], errors='coerce')

# Lista de columnas numéricas que queremos convertir a sus tipos adecuados
numeric_columns = ['AGE', 'SALARY', 'LEAVES USED', 'LEAVES REMAINING', 'RATINGS', 'PAST EXP']

# Usar un bucle para convertir todas las columnas numéricas
for col in numeric_columns:
    df_current[col] = pd.to_numeric(df_current[col], errors='coerce').astype('Int64')

# Imprimir los tipos de datos para asegurarse de que todo se haya convertido correctamente
print("Conversion finalizada")
display(df_current.dtypes)
print("-"*50)
display(df_current)
```

```python
info_general_df(df_current)
```

```python
# Procesamiento de datos faltantes.
data_fillna = {
    'LAST NAME': "Desconocido",  # Asignamos una cadena para los nombres faltantes

    # Generar fechas aleatorias para las columnas de fechas
    'DOJ': pd.to_datetime(np.random.uniform(df_current['DOJ'].min().value, df_current['DOJ'].max().value)).normalize(),
    'CURRENT DATE': pd.to_datetime(np.random.uniform(df_current['CURRENT DATE'].min().value, df_current['CURRENT DATE'].max().value)),

    # Usar la media para completar valores numéricos faltantes
    'AGE': int(df_current['AGE'].mean()),  # Convertimos a entero usando int()
    'LEAVES USED': int(df_current['LEAVES USED'].mean()),  # Similar, convertimos a entero
    'LEAVES REMAINING': int(df_current['LEAVES REMAINING'].mean()),  # Convertimos la media a entero
    'RATINGS': int(df_current['RATINGS'].mean())  # Convertimos la media a entero
}

# Usar fillna con el diccionario
df_current.fillna(data_fillna, inplace=True)
# Verificar la cantidad de nulos restantes
print(df_current.isnull().sum())
```

## **Calcular metricas y rangos de datos**

```python
# Función para calcular antigüedad en años y días restantes
def calcular_antiguedad(row):
    diferencia = row['current_date'] - row['date_joining']
    antiguedad_dias = diferencia.days
    antiguedad_anios = antiguedad_dias // 365  # Aproximación por años completos
    dias_restantes = antiguedad_dias % 365  # Días restantes después de los años completos
    return pd.Series([antiguedad_anios, dias_restantes])

# Aplicar la función usando apply
df_metrics = df_current.copy()
df_metrics[['tenure_years', 'remaining_days']] = df_metrics.apply(calcular_antiguedad, axis=1)

# Mostrar el DataFrame resultante
df_metrics[['tenure_years', 'remaining_days']]
```

```python
# aplicamos la funcion de map, para generar un rango de edades (age)
def age_range(age):
    if 20 <= age <= 25:
        return '20-25'
    elif 26 <= age <= 30:
        return '26-30'
    elif 31 <= age <= 35:
        return '31-35'
    elif 36 <= age <= 40:
        return '36-40'
    elif 41 <= age <= 45:
        return '41-45'
    else:
        return 'Unknown'

df_metrics['age_range'] = df_metrics['age'].map(age_range)
df_metrics['age_range']
```

```python
info_general_df(df_metrics)
```

```python
# Filtramos informacion de interes con respecto a rangos de edades '20-25'
df_metrics_20_25 = df_metrics[df_metrics['age_range'] == '41-45']
df_metrics_20_25.sort_values(by='salary', ascending=False)
```

```python
# Agrupar por experiencia previa (past_exp) y calcular el salario promedio
experiencia_vs_salario = df_metrics.groupby('past_exp')['salary'].mean()
experiencia_vs_salario
```

```python
# Agrupar por departamento y calcular el salario promedio
salario_por_departamento = df_metrics.groupby('unit')['salary'].mean()
salario_por_departamento
```
